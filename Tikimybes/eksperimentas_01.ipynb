```python
#@title Importuoti paketus

# biblioteka generuoti atsitiktiniams skaičiams
import random

# interaktyvių grafikų piešimas
import plotly.graph_objects as go
import plotly.io as pio

# nustatyti temą
pio.templates.default = "seaborn"
```

## I. Apskaičiuoti &pi; vertę

&pi; vertė yra iracionalus skaičius (`3.141592653589793238...`) , t.y. .

https://en.wikipedia.org/wiki/Pi

```python
#@title Simuliacija

%%time
# bandimų skaičius
TRIALS = 1000 # @param {type:"integer"}

# konstanta sekti, kiek kartų taškas pakliuvo į apskritimą
circle_points = 0

# PI vertės
PI_values = list()

# atlikti N kartų (N = TRIALS) simulaicijas
for i in range(TRIALS):
  # sugeneruok atsitiktinį skaičių x ir y kordinatėms
  rand_x = random.uniform(-1, 1)
  rand_y = random.uniform(-1, 1)

  # apskaičiuoti radiusą/atstumą nuo centro
  origin_dist = rand_x**2 + rand_y**2

  # patikrinti ar taškas (x, y) pakliuvo į apskritimą
  if origin_dist <= 1:
    circle_points += 1

  # paskaičiuoti PI vertę ir ją išsaugoti
  PI_values.append(4 * circle_points / (i + 1))


# atspausinti paskutinį simuliacijų rezultatą
print(f"Po {TRIALS} bandimų, apskaičiuota PI vertė yra {PI_values[-1]:.5f}")
print("Tiksli PI vertė 5 skaičiai po kablelio yra 3.14159.")
_paklaida = (PI_values[-1]-3.14159)/3.14159*100
print(f"Skirtumas {PI_values[-1]-3.14159:.5f} arba {_paklaida:.2f}% paklaida.\n")
```

```python
#@title Simuliacijų vizualizacija
# sukurti grafiko objektą
fig = go.Figure()

# sugeneruoti x ir y ašies vertes
x_values = [_+1 for _ in range(TRIALS)]
y_values = PI_values

# išfiltruoti taškus tarpinias simuliacijų vertes, kad grafikas greit nusipieštų
if TRIALS > 500:
  # nustatyti žingsnį
  step = (TRIALS-500)//100

  # visalaik paimti pirmus 500 bandymų
  x_values = x_values[:500] + x_values[499+step::step]
  y_values = y_values[:500] + y_values[499+step::step]

# nupiešti liniją su taškais
fig.add_trace(go.Scatter(x=x_values, y=y_values,
                         mode='lines+markers', name='Simuliacijos'))

# pridėti teorinę vertę
fig.add_trace(go.Scatter(x=[0, TRIALS], y=[3.1415926, 3.1415926],
                         mode='lines', name='Teorinė'))

# atnaujinti paveiksliuko parametrus
fig.update_layout(title=f'Monte Carlo simuliacijos (N={TRIALS})',
                   xaxis_title='Bandimų sk.',
                   yaxis_title='PI vertė')

# gražinti grafiką į konsolę
fig.show()
```
